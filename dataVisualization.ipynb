```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
data['Outcome'] = data['Outcome'].apply(lambda x: 'Successful' if 'True' in str(x) else 'Unsuccessful')
data.head()
```

```python
X_axis = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_3.csv')
X_axis.head()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Outcome', data=data, hue="Outcome", palette="Blues", alpha=0.9, legend=False)
plt.title('Overall Launch Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='LaunchSite', hue='Outcome', palette="Blues", data=data)
plt.title('Launch Outcomes Based on Launch Site')
plt.xlabel('Launch Site')
plt.ylabel('Count')
plt.legend(title='Outcome', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Reused', hue='Outcome', data=data, edgecolor='black', linewidth=0.5, palette= "Blues")
plt.title('Launch Outcomes Based on Reused Booster')
plt.xlabel('Booster Reused')
plt.ylabel('Count')
plt.legend(title='Outcome', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Legs', hue='Outcome', data=data, palette = "Blues")
plt.title('Launch Outcome Based on Reused Legs')
plt.xlabel('Legs Reused')
plt.ylabel('Count')
plt.legend(title='Outcome', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Outcome', y='PayloadMass', hue = "Outcome", data=data, palette="Blues")
plt.title('Payload Mass vs Outcome')
plt.xlabel('Outcome')
plt.ylabel('Payload Mass (Kilograms)')
plt.show()
```
